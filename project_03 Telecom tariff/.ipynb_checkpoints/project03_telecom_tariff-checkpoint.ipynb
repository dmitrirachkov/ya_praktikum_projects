```python
def pivot_by_userid_monthly(df,col,func,new_df):
    '''Создаёт сводную таблицу по нужной колонке с нужной функцией (df,col,func,new_df)'''
    new_df = df.pivot_table(
        index=['user_id','month_year'], values=[col], aggfunc=func).fillna(0).reset_index()
    return new_df

def to_month_year(df,date_col):
    '''Создаёт колонку month_year в заданной таблице по заданной колонке с датой (df,date_col)'''
    df['month_year']= df[date_col].values.astype('datetime64[M]')
    return
```

### Задача 2.4 НАЙДЁМ помесячную выручку с каждого пользователя 

--> Вычтите бесплатный лимит из суммарного количества звонков, сообщений и - интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану)


```python
# Для начала избавимся от пропусков
# summary.info()
```

```python
summary = summary.fillna(0)
summary.info()
```

```python
summary['calls_revenue'] = (summary['calls_count'] - summary['minutes_included'])*summary['rub_per_minute']
summary['message_revenue'] = (summary['message_count'] - summary['messages_included'])*summary['rub_per_message']
summary['internet_revenue'] = (((summary['mb_used_ceil'] - summary['mb_per_month_included'])/1024)
                               .apply(np.ceil))*summary['rub_per_gb']

#----> При сложении нужно НЕ УЧИТЫВАТЬ минусы!!!
summary['calls_revenue_final'] = summary['calls_revenue']
summary.loc[summary['calls_revenue'] < 0, 'calls_revenue_final'] = 0
summary['message_revenue_final'] = summary['message_revenue']
summary.loc[summary['message_revenue'] < 0, 'message_revenue_final'] = 0
summary['internet_revenue_final'] = summary['internet_revenue']
summary.loc[summary['internet_revenue'] < 0, 'internet_revenue_final'] = 0

summary['user_revenue'] = (summary['rub_monthly_fee'] + summary['calls_revenue_final'] + 
                           summary['message_revenue_final'] + summary['internet_revenue_final'])
summary.sample(10)
```

```python
summary.groupby('month')['reg_month'].count().plot(title='Суммарное число пользователей', figsize=(12,4), grid=True)
```

```python
summary.groupby('month')['user_revenue'].sum().plot(title='Суммарная выручка', figsize=(12,4), grid=True)
```

Из графиков видно, что период август-сентябрь были "проблемным" т.к. темпы роста выручки снизились относительно темпов увеличения числа клиентов

```python
# посчитаем дисперсию
# sigma**2 = sum(mean-xi)**2 / n

# Посчитаем дисперсию (variance) стандартным методом библиотеки numpy
variance_calls = np.var(mean_data['calls_duration_ceil']) #     8'518.85
variance_internet = np.var(mean_data['mb_used_ceil'])     # 9'768'012.71
variance_messages = np.var(mean_data['message_count'])    #       118.53
print(variance_calls, variance_internet, variance_messages) # странно, почему-то не работает форматирование 
                                                            # вывода 2 знаков после запятой для float
```

```python
# Посчитаем стандартное отклонение (standard_deviation) методом numpy
std_dev_calls = np.std(mean_data['calls_duration_ceil']) #    92.29
std_dev_internet = np.std(mean_data['mb_used_ceil'])     # 3'125.38
std_dev_messages = np.std(mean_data['message_count'])    #    10.88
print(std_dev_calls, std_dev_internet, std_dev_messages)
```

```python
# проверим то же самое через вычисление квадрата
print(np.sqrt(variance_calls), np.sqrt(variance_internet), np.sqrt(variance_messages)) # --> Совпадает! =) 
```

## Шаг 4. Проверка гипотез
 - Гипотеза 1: средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
 - Гипотеза 2: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;
 - Пороговое значение alpha задайте самостоятельно.

<u>Поясните:</u>
 - как вы формулировали нулевую и альтернативную гипотезы;
 - какой критерий использовали для проверки гипотез и почему.

```python
# Проверим гипотезу 1
# Нулевая гипотеза: Средняя выручка пользователей тарифов "Ультра" и "Смарт" одинаковы.
# Альтернативная гипотеза: Средняя выручка пользователей тарифов "Ультра" и "Смарт" НЕ одинаковы.

revenue_ultra = summary[summary['tariff_name'] == 'ultra']['user_revenue'].to_list()
revenue_smart = summary[summary['tariff_name'] == 'smart']['user_revenue'].to_list()

alpha=.05 # критический уровень статичтической значимости

result_hyp1 = st.ttest_ind(revenue_ultra,revenue_smart)
result_hyp1
```

В нашем случае pvalue=1.87 , что больше alpha=0.05, следовательно "нулевую гипотезу отвергнуть не удалось", т.е. средняя выручка пользователей двух тарифов может быть одинакова

```python
# Проверим вручную
summary[summary['tariff_name'] == 'ultra']['user_revenue'].mean() # 2070.94
summary[summary['tariff_name'] == 'smart']['user_revenue'].mean() # 1170.53

# На первый взгляд значения отличаются почти в 2 раза. 
#--> Но теория говорит, что подобный разброс может быть получен и случайно!!!
```

---

```python
# Проверим гипотезу 2
# Нулевая гипотеза: Средняя выручка пользователей из Москвы равна ср. выручке пользователей из других регионов.
# Альтернативная гипотеза: Ср. выручка пользователей из Москвы НЕ равна ср. выручке пользователей из других регионов.

# Для начала необходимо добавить в summary информацию по городу каждого пользователя
```

```python
u_city = u[['user_id', 'city']].copy()
u_city.rename(columns={'city':'user_city'}, inplace=True)
u_city.head(3)
```

```python
if 'user_city' in summary.columns:
    pass
else:
    summary = summary.merge(u_city, on='user_id')

summary = summary[['user_id', 'user_city', 'reg_month', 'month', 'calls_count', 'calls_duration_ceil',(
    'calls_duration'), 'message_count', 'mb_used_ceil', 'mb_used','tariff_name', 'churn_date',(
    'messages_included'),'mb_per_month_included', 'minutes_included', 'rub_monthly_fee',(
    'rub_per_gb'), 'rub_per_message', 'rub_per_minute', 'calls_revenue','message_revenue',(
    'internet_revenue'), 'calls_revenue_final','message_revenue_final', 'internet_revenue_final','user_revenue']]
summary.sample()
```

```python
revenue_msk = summary[summary['user_city'] == 'Москва']['user_revenue'].to_list()
revenue_rus = summary[summary['user_city'] != 'Москва']['user_revenue'].to_list()

alpha=.05 # критический уровень статичтической значимости

result_hyp2 = st.ttest_ind(revenue_msk,revenue_rus)
result_hyp2
```

В нашем случае pvalue=0.28, что больше alpha=0.05, следовательно "нулевую гипотезу отвергнуть не удалось", т.е. средняя выручка пользователей из Москвы может быть одинакова ср. выручке пользователей из других регионов.

```python
# Проверим вручную
summary[summary['user_city'] == 'Москва']['user_revenue'].mean() # 1476.68
summary[summary['user_city'] != 'Москва']['user_revenue'].mean() # 1439.53

# Ручная проверка показала, что средние в МСК и остальных городов - почти совпадают!
```

Пояснения:
1. Нулевую гипотезу формулируем всегда утвердительно! Альтернативную - на отрицании нулевой гипотезы.
2. Для проверки использовал pvalue, получаемый методом ttest_ind (для сравнения двух совокупностей или их выборок). 

Если pvalue больше заданного критического уровня стат. значимости alpha, то нулевую гипотезу отвергать нельзя!
